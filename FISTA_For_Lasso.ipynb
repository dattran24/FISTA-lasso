```
#Generate Data################################OK########################
#Input contains 
#1. N, dim (which are the length of b and the dimension of x in the paper), 
#2. mu (which is lambda in the paper), 
#3. A,b 
import numpy as np
N = 784
dim = 784
mu = 0.00001
np.random.seed(50)
w = np.matrix(np.random.multivariate_normal([0.0]*dim, np.eye(dim))).T
A_gen = np.matrix(np.random.multivariate_normal([0.0]*dim, np.eye(dim), size = N))
b_com = A_gen*w
b=np.zeros(N)
for i in range(len(b)):
  b[i]=b_com[i][0]
A=np.array(A_gen)
```

```
#Tim gia tri
#Finding value of F(x)=1/2*||Ax-b||^2+ \mu ||x||_1
B=np.array(A)
def giatri(x):
  gt=0
  for i in range(0,dim):
    gt+=0.5*(np.matmul(B,x)-b)[i]**2+mu*abs(x[i])
  return gt
import decimal
import numpy as np 
from numpy import linalg as LA
import math
#Finding Lipschitz constant by finding maximum eigenvalue of A^TA
def eigenmax(A): 
  A1=np.array(A)
  A2=A1.transpose()
  Q=np.matmul(A2,A1)
  w,v=LA.eig(Q)
  e=max(w)
  return e 
#Calculate Prox function
def Prox(y,mu,gamma):
  m=np.zeros(len(y))
  for i in range(0,len(y)):
    if y[i]>=mu*gamma:
      m[i]=y[i]-mu*gamma
    if y[i]<=-mu*gamma:
      m[i]=y[i]+mu*gamma
    if (y[i]<mu*gamma) and (y[i]>-mu*gamma):
      m[i]=0
  return m
def PL(y):
  return Prox(y-1/eigenmax(A)*np.matmul(np.array(A).transpose(),np.matmul(np.array(A),y)-np.array(b)),mu,1/eigenmax(A)) 
iter=3000
y=np.ones(784)
x=y
step=0
t=1
import time
u=[]
start_time = time.time()
value=[]
#Folowwing is a condition for stop algorithm
#We can change it into while step<iter:
while giatri(x)>0.006141384 and (float(time.time() - start_time)<4000):
  if step%10==0:
    print(giatri(x))
  x0=x
  x=PL(y)
  t0=t
  t=0.5*(1+math.sqrt(1+4*t**2))
  y=x+((t0-1)/(t))*(x-x0)
  u.append(float(time.time() - start_time))
  value.append(giatri(x))
  step+=1
```
